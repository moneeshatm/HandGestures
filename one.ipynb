```python
import mediapipe as mp
import cv2
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
data=mp_hands.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(img,x,mp_hands.HAND_CONNECTIONS,mp_drawing.DrawingSpec(color=(255,0,0),thickness=5,circle_radius=5),mp_drawing.DrawingSpec(color=(255,0,0),thickness=5))



    cv2.imshow("Hand",img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
