```python
import mediapipe as mp
import cv2
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
data=mp_hands.Hands(max_num_hands=1)
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    # print(result.multi_hand_landmarks)
    tipid=[4,8,12,16,20]
    lmlist=[]
    if result.multi_hand_landmarks:
        for handlms in result.multi_hand_landmarks:
            for id, lm in enumerate(handlms.landmark):
                # print(id,lm)#to print index of landmarks
                cx=lm.x #collects value of x co-ordinate only
                cy=lm.y
                lmlist.append([id,cx,cy])
                # print(lmlist)
                if len(lmlist) !=0 and len(lmlist)==21:
                    fingercount=[]
                    #thumb
                    if lmlist[12][1]<lmlist[20][1]:#to detect which hand
                        if lmlist[4][1]>lmlist[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    else:
                        if lmlist[4][1]<lmlist[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)

                        #all fingers except thumb
                    for i in range(1,5): #i=0, (5)->includes thumb
                        if lmlist[tipid[i]][2]>lmlist[tipid[i]-2][2]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    # print(fingercount)
                    if len(fingercount)!=0:
                        fingercount1=fingercount.count(1)
                    # print(fingercount1)
                    cv2.putText(img,"Count: "+str(fingercount1),(35,400),cv2.FONT_HERSHEY_COMPLEX,3,(255,0,0),3)
            mp_drawing.draw_landmarks(img,handlms,mp_hands.HAND_CONNECTIONS)



    cv2.imshow("Finger Count",img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
